```python
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline

model = AutoModelForCausalLM.from_pretrained("duyan2803/llama-2-7b-chat-finetune", device_map="auto", load_in_8bit=True, llm_int8_enable_fp32_cpu_offload=True)
tokenizer = AutoTokenizer.from_pretrained("duyan2803/llama-2-7b-chat-finetune")
chat = pipeline("text-generation", model=model, tokenizer=tokenizer)
```

```python
prompt = "Ngày bác Hồ ra đi tìm đường cứu nước là ngày nào?"
result = chat(prompt, max_new_tokens=200)
print(result)
```
